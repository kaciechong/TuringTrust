```python
# Objective: To analyze the total number of drives that have been wiped successfully more than once over time, grouped by month.

# Hypothesis: Number of drives wiped unnecessarily will decrease over time after the implementation of new guards or measures (since Aug 2023)

pipeline = [
    # 1. Unwind the "erasures" array field to create a document for each element in the array
    {
        "$unwind": {
            "path": "$erasures",
            "preserveNullAndEmptyArrays": False
        }
    },
    
    # 2. Filter documents based on conditions for "erasures" fields and "system.manufacturer"
    {
        "$match": {
            "erasures.serial": { "$exists": True, "$ne": None },
            "erasures.state": "Successful",
            "erasures.interface_type": { "$ne": "USB" },
            "system.manufacturer": { "$ne": "Apple, Inc" }
        }
    },
    
    # 3. Group by "erasures.serial" and "date" to find the most recent wipe date per serial,
    #    retaining the original document _id for future reference
    {
        "$group": {
            "_id": { "serial": "$erasures.serial", "date": "$date" },
            "mostRecentDate": { "$max": "$date" },
            "originalId": { "$first": "$_id" }  # Retain the original document _id
        }
    },
    
    # 4. Group by "serial" to calculate the wipe count and collect the most recent wipe date
    #    along with the original document _id and unique serials for each document
    {
        "$group": {
            "_id": "$_id.serial",
            "wipeCount": { "$sum": 1 },
            "mostRecentDate": { "$max": "$mostRecentDate" },
            "originalId": { "$first": "$originalId" },
            "totalDrives": { "$addToSet": "$_id.serial" }  # Collect unique serials
        }
    },
    
    # 5. Group by the month and year of the most recent wipe date to count drives wiped in each month
    #    and store the original documents for further processing
    {
        "$group": {
            "_id": { "month": { "$month": "$mostRecentDate" }, "year": { "$year": "$mostRecentDate" } },
            "drivesInMonth": { "$sum": 1 },
            "docs": { "$push": "$$ROOT" }  # Keep original documents for further processing
        }
    },
    
    # 6. Unwind the grouped documents for further processing
    {
        "$unwind": "$docs"
    },
    
    # 7. Replace the root document with merged objects, including the drive count per month
    {
        "$replaceRoot": { "newRoot": { "$mergeObjects": [ "$docs", { "drivesInMonth": "$drivesInMonth" } ] } }
    },
    
    # 8. Project specific fields to return, including calculations of total and adjusted drives
    {
        "$project": {
            "_id": "$originalId",  # Return the original document _id
            "wipeCount": 1,
            "mostRecentDate": 1,
            "serial": "$_id",  # Return the serial as its own field
            "totalDrives": { "$size": "$totalDrives" },
            "adjustedTotalDrives": { "$divide": [{ "$size": "$totalDrives" }, "$drivesInMonth" ] }  # Adjusted total drives
        }
    },
    
    # 9. Filter out documents where the wipe count is not greater than 1
    {
        "$match": {
            "wipeCount": { "$gt": 1 }
        }
    }
]

results = list(all_reports_collection.aggregate(pipeline))
df = pd.DataFrame(results)
display(df)
```

```python
# Ensure 'Most Recent Date' is in datetime format
df['mostRecentDate'] = pd.to_datetime(df['mostRecentDate'])

# Group by 'Most Recent Date' (month-year) and sum 'Adjusted Total Drives' for each month
df_grouped_by_month = df.groupby(df['mostRecentDate'].dt.to_period('M')).agg({'adjustedTotalDrives': 'sum'}).reset_index()

# Exclude the last month from the grouped data
df_grouped_by_month = df_grouped_by_month[:-1]

# Plot the bar graph
plt.figure(figsize=(12, 6))  # Adjust the figure size for better clarity
plt.bar(df_grouped_by_month['mostRecentDate'].astype(str), df_grouped_by_month['adjustedTotalDrives'], color='skyblue')

# Add labels and title to the graph
plt.xlabel('Month')
plt.ylabel('Total Adjusted Drives')
plt.title('Percentage of Drives Successfully Wiped Multiple Times (Unnecessary Wipes / Total Wipes)')
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.tight_layout()  # Ensure everything fits within the plot area

plt.savefig('drives_wiped_multiple_times_trend.png', dpi=300)  # You can specify the format here (e.g., .png, .jpg)
plt.show()
```
