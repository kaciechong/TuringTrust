```python
import pandas as pd

df = pd.read_csv("/Users/kaciechong/CURRENT.csv")

# Filter for rows where rigWipeStatuses Failed
failed_wipes = df[df['rigWipeStatuses'] == 'failed']
failure_counts = failed_wipes['failureMessages'].value_counts()
print(failure_counts)
```

```python
# Group statistics for each failure message
stats = df.groupby("failureMessages").agg(
    total_occurrences=("rigWipeStatuses", "count"),  # Total occurrences of each failure message
    failed_occurrences=("rigWipeStatuses", lambda x: (x == "failed").sum()),  # Count of Failed rig wipes -- REVIEW LAMBDA!!
)

# Calculate failure rate
stats["failure_rate"] = (stats["failed_occurrences"] / stats["total_occurrences"]) * 100

stats = stats.sort_values(by="failure_rate", ascending=False)
stats
```

```python
df2 = pd.read_csv("/Users/kaciechong/CURRENT2.csv")

stats2 = df2.groupby("failureMessages").agg(
    total_occurrences2=("rigWipeStatuses", "count"),  # Total occurrences of each failure message
    failed_occurrences2=("rigWipeStatuses", lambda x: x.str.lower().eq("failed").sum()),  # Count of Failed rig wipes
)

# Calculate failure rate
stats2["failure_rate"] = (stats2["failed_occurrences2"] / stats2["total_occurrences2"]) * 100

# Sort by failure rate in descending order
stats2 = stats2.sort_values(by="failure_rate", ascending=False)

stats2
```
